```python
import requests
import time
import pandas as pd

url = "https://api.brightdata.com/datasets/v3/snapshot/s_mb6lmemy1s63yzlbrs"
headers = {
	"Authorization": "Bearer 7effc080-0480-4f16-b916-fb435fc0a508",
}
params = {
	"format": "json",
}

response = requests.get(url, headers=headers, params=params)
response_json = response.json()

while isinstance(response_json, dict) and response_json.get("status") == "running":
    time.sleep(300)
    response = requests.get(url, headers=headers, params=params)
    response_json = response.json()

keyword_search = pd.DataFrame(response_json)
keyword_search.to_csv("keyword_search.csv", index=False)
```
